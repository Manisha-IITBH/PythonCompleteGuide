1. Load and Inspect Data

- Read health_data.csv using pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('health_data.csv')
print(df)
print("\n\n\nDataset Info:")
print(df.info())
```

2. Line Chart â€“ Fitness Progress

- Plot monthly Steps and Calories Burned.

- Highlight months with the highest activity.

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Plot Steps and Calories Burned
df.plot(kind='line', x='Month', y='Steps', ax=ax, label='Steps', color='blue')
df.plot(kind='line', x='Month', y='Calories_Burned', ax=ax, label='Calories Burned', color='red')

# max values
max_steps = df['Steps'].max()
max_calories = df['Calories_Burned'].max()

# months where max values occur
max_steps_month = df.loc[df['Steps'] == max_steps, 'Month'].values[0]
max_calories_month = df.loc[df['Calories_Burned'] == max_calories, 'Month'].values[0]

#max_steps_months = df.loc[df['Steps'] == max_steps, 'Month'].tolist() in case we want to convert to list

# Highlight max activity with vertical lines
ax.axvline(x=max_steps_month, color='black', linestyle='--', label=f'Max Steps: {max_steps}')
ax.axvline(x=max_calories_month, color='orange', linestyle='--', label=f'Max Calories: {max_calories}')

# Set Lables
ax.set_xlabel('Month')
ax.set_ylabel('Activity')
ax.set_title('Monthly Fitness Progress')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)

# # Adjust layout and show
# plt.tight_layout()
# plt.show()
```

3. Bar Chart â€“ Sleep Analysis

- Plot average Sleep Hours per Month using a bar chart.

- Add a horizontal line at 7 hours as a reference for healthy sleep.

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Plot average Sleep Hours per Month
df.plot(kind='bar', x='Month', y='Sleep_Hours', ax=ax, label='Sleep Hours', color='skyblue')
ax.axhline(y=7, color='red', linestyle='--', label='Healthy Sleep (7 hours)')

# Set Lables
ax.set_xlabel('Month')
ax.set_ylim(0, 9)
ax.set_ylabel('Sleep Hours')
ax.set_title('Average Sleep Hours per Month')
ax.legend(loc=0)
ax.grid(True, linestyle='--', alpha=0.5)
```

4. Scatter Plot â€“ Heart Rate vs Sleep

- Plot Avg_Heart_Rate against Sleep_Hours.

- Use color coding for Steps that month.

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Create scatter plot with color coding based on Steps
scatter = ax.scatter(df['Avg_Heart_Rate'], df['Sleep_Hours'],
                     c=df['Steps'], cmap='winter', s=100, edgecolors='black')

# Add color bar to explain the Steps color coding
cbar = plt.colorbar(scatter)
cbar.set_label('Steps')

# Set labels and title
ax.set_xlabel('Average Heart Rate')
ax.set_ylabel('Sleep Hours')
ax.set_ylim(0, 8)
ax.set_title('Average Heart Rate vs Sleep Hours')

# Add grid and legend
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(['Sleep Hours'], loc='best')
```

5. Pie Chart â€“ Activity Contribution

- Compare total yearly Steps vs total Calories Burned (scaled).

- Show percentage contribution.

```python
# Calculate total steps and total calories burned
total_steps = df['Steps'].sum()
total_calories = df['Calories_Burned'].sum()
print(total_steps)
print(total_calories)
total = total_steps + total_calories
steps_percentage = total_steps / total
calories_percentage = total_calories / total
print(steps_percentage)
print(calories_percentage)

# Values and labels for the pie chart
values = [steps_percentage, calories_percentage]
labels = ['Total Steps', 'Calories Burned']
colors = ['skyblue', 'orange']

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(values, labels=labels, autopct='%1.1f%%',colors=colors)
#ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Activity Contribution')
```

Subplots (2x2 Layout)

- Combine line, bar, scatter, and pie charts in one figure.

- Add the main title: 'Health & Fitness Analysis (2024)'.

```python
# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line Chart
ax = axs[0, 0]
df.plot(kind='line', x='Month', y='Steps', ax=ax, label='Steps', color='blue')
df.plot(kind='line', x='Month', y='Calories_Burned', ax=ax, label='Calories Burned', color='red')


max_steps = df['Steps'].max()
max_calories = df['Calories_Burned'].max()
max_steps_month = df.loc[df['Steps'] == max_steps, 'Month'].values[0]
max_calories_month = df.loc[df['Calories_Burned'] == max_calories, 'Month'].values[0]
ax.axvline(x=max_steps_month, color='black', linestyle='--', label=f'Max Steps')
ax.axvline(x=max_calories_month, color='orange', linestyle='--', label=f'Max Calories')

ax.set_xlabel('Month')
ax.set_ylabel('Activity')
ax.set_title('Monthly Fitness Progress')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)

# Bar Chart
ax = axs[0, 1]
df.plot(kind='bar', x='Month', y='Sleep_Hours', ax=ax, color='skyblue')
ax.axhline(y=7, color='red', linestyle='--', label='Healthy Sleep (7 hours)')
ax.set_xlabel('Month')
ax.set_ylim(0, 9)
ax.set_ylabel('Sleep Hours')
ax.set_title('Average Sleep Hours per Month')
ax.legend(loc='best')
ax.grid(True, linestyle='--', alpha=0.5)

# Scatter Plot
ax = axs[1, 0]
scatter = ax.scatter(df['Avg_Heart_Rate'], df['Sleep_Hours'],
                     c=df['Steps'], cmap='winter', s=100, edgecolors='black')
cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('Steps')
ax.set_xlabel('Average Heart Rate')
ax.set_ylabel('Sleep Hours')
ax.set_ylim(0, 8)
ax.set_title('Average Heart Rate vs Sleep Hours')
ax.grid(True, linestyle='--', alpha=0.5)

# Pie Chart
ax = axs[1, 1]
total_steps = df['Steps'].sum()
total_calories = df['Calories_Burned'].sum()
total = total_steps + total_calories
steps_percentage = total_steps / total
calories_percentage = total_calories / total
values = [steps_percentage, calories_percentage]
labels = ['Total Steps', 'Calories Burned']
colors = ['skyblue', 'orange']
ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
ax.set_title('Activity Contribution')


fig.suptitle('Health & Fitness Analysis (2024)', fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # leave space for suptitle
#plt.show()
```
