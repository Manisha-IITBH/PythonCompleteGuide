```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = sns.load_dataset('iris')
iris_df.head()
```

1. Plot the relationship between PetalLengthCm and PetalWidthCm using a 2D
scatter plot.

```python
sns.scatterplot(data=iris_df, x='petal_length', y='petal_width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Relationship between Petal Length and Petal Width')
```

```python

# 2D scatter plot colored by species
sns.scatterplot(data=iris_df, x='petal_length', y='petal_width', hue='species', palette='Set1')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Relationship between Petal Length and Petal Width by Species')

```

2. Now, plot a 3D scatter plot by adding SepalLengthCm as the third dimension.

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')

# 3D scatter plot
scatter = ax.scatter(
    iris_df['petal_length'],
    iris_df['petal_width'],
    iris_df['sepal_length'],
    c=iris_df['species'].astype('category').cat.codes,  # color by species
    cmap='viridis',
    s=50
)

# Labeling the axes
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_zlabel('Sepal Length (cm)')
ax.set_title('3D Scatter Plot of Iris Dataset')


legend1 = ax.legend(*scatter.legend_elements(), title="Species")
ax.add_artist(legend1)
plt.show()
```

# Which plot (2D or 3D) is more effective in helping you distinguish?


The 2D scatter plot is more effective for quickly distinguishing species because two variables already provide clear separation.
