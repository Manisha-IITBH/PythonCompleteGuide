Problem Set-1:

Download the heart disease dataset from the Canvas and perform the following
tasks. The dataset contains the following columns.

Dataset
- age: Patient’s age
- sex: Gender (1 = male, 0 = female)
- cp: Chest pain type (categorical: 0–3)
- trestbps: Resting blood pressure
- chol: Serum cholesterol (mg/dl)
- thalach: Maximum heart rate achieved
- target: Heart disease presence (1 = yes, 0 = no)


  
Task 1: Pairplot
- Create a pairplot for the columns ['age', 'chol', 'thalach', 'target'].
- Use hue='target' to differentiate between patients with and without heart disease.
- Write a short explanation of any visible separation or clustering.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("heart.csv")
df.head()
```

```python
sns.pairplot(df[['age', 'chol', 'thalach', 'target']], hue="target")
plt.show()
```

## Task 2: Boxplot
- Draw a boxplot of trestbps (resting blood pressure) grouped by target.
- Add another boxplot for chol grouped by sex.
- Interpret whether there are differences in distributions.

```python
plt.figure(figsize=( 8, 5))
sns.boxplot(x= "target", y="trestbps",data=df)
plt.title("Boxplot of trestbps (resting blood pressure) grouped by target")
plt.xlabel("Target" )
plt.ylabel("Resting blood pressure" )
#plt.legend(title= "Sex")
plt.show()
```

```python
plt.figure(figsize=( 8, 5))
sns.boxplot(x= "target", y="chol",data=df, hue= "sex")
plt.title("Boxplot for chol grouped by sex")
plt.xlabel("Target" )
plt.ylabel("chol" )
plt.legend(title= "Sex")
plt.show()
```

## Task 3: Violin Plot

- Plot a violin plot of thalach (max heart rate) split by target.
- Add another violin plot for age split by sex.
- Compare the spread and medians across groups.

```python
sns.violinplot(x= 'target', y='thalach',data=df)
# Add title and labels
plt.title('Violin plot of thalach (max heart rate) split by target.')
plt.xlabel('target')
plt.ylabel('thalach' )
# Show the plot
plt.show()
```

```python
sns.violinplot(x= 'target', y='age',data=df)
# Add title and labels
plt.title('violin plot for age split by sex')
plt.xlabel('target')
plt.ylabel('age' )
# Show the plot
plt.show()
```

## Task 4: Heatmap

- Compute the correlation matrix for all numerical features.
- Plot a heatmap with annotations.
- Identify which variables are strongly correlated with the target.

```python
plt.figure(figsize=(10, 10))
corr=df.corr()
sns.heatmap(corr, annot= True, fmt = "0.1f",cmap='YlGnBu', linewidths=0.9)
plt.title('Correlation matrix for all numerical features' )
plt.show()
```

## Task 5: Jointplot

- Create a jointplot of age vs thalach with hue='target'.
- Try both kind='scatter' and kind='kde'.
- Explain whether younger or older patients tend to achieve higher max heart rates.



```python
plt.figure(figsize=(10, 10))
sns.jointplot(x="age", y="thalach", data=df, kind="scatter", color="blue", hue="target")
plt.title('jointplot of age vs thalach' )
plt.show()
```

## Task 6: Barplot

- Create a barplot of cp (chest pain type) vs average thalach, with hue='target'.
- Create another barplot for sex vs average chol, with hue='target'.
- Discuss differences across groups.

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="cp", y="thalach", data=df, hue="target")
plt.title('Barplot of cp (chest pain type) vs average thalach' )
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="sex", y="chol", data=df, hue="target")
plt.title('Barplot of cp (chest pain type) vs average thalach' )
plt.show()
```
